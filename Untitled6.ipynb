## 1. Fungsi-fungsi Utilitas
Semua fungsi untuk load data, cleaning, dan agregasi didefinisikan di sini.

```python
import json

def load_data_from_file(nama_file):
    """Membaca file JSON (list of objects) dan mengembalikannya."""
    with open(nama_file, 'r') as f:
        data = json.load(f)
    return data

def clean_data(transactions):
    """
    Membersihkan data:
    - Hapus transaksi dengan price negatif atau None
    - Hapus jika item_name kosong
    - Ubah product_category menjadi lowercase dan strip spasi
    """
    cleaned = []
    for t in transactions:
        # Filter price
        price = t.get('price')
        if price is None or price < 0:
            continue

        # Filter item_name
        if not t.get('item_name'):
            continue

        # Salin dictionary
        cleaned_t = t.copy()

        # Bersihkan kategori
        kategori = cleaned_t.get('product_category', '')
        if isinstance(kategori, str):
            cleaned_t['product_category'] = kategori.lower().strip()
        else:
            cleaned_t['product_category'] = str(kategori).lower().strip()

        cleaned.append(cleaned_t)
    return cleaned

def get_unique_customers(cleaned_data):
    """Mengembalikan set ID pelanggan unik."""
    return {t['customer_id'] for t in cleaned_data if 'customer_id' in t}

def aggregate_revenue_per_category(cleaned_data):
    """
    Menghitung total revenue per kategori.
    Revenue = price (karena tidak ada quantity)
    """
    revenue = {}
    for t in cleaned_data:
        kategori = t.get('product_category', 'unknown')
        price = t.get('price', 0)
        revenue[kategori] = revenue.get(kategori, 0) + price
    return revenue
```

## 2. Eksekusi Pipeline
Memanggil fungsi-fungsi secara berurutan.

```python
# Nama file yang sudah diupload
file_json = "transaction_log.json"   # ganti jika nama file berbeda

Load data
print("Membaca file...")
data_mentah = load_data_from_file(file_json)
print(f"Jumlah data mentah: {len(data_mentah)}")

# 1. Bersihkan data
print("\nMembersihkan data...")
cleaned_data = clean_data(data_mentah)
print(f"Jumlah data setelah pembersihan: {len(cleaned_data)}")
print(f"Jumlah data yang dibuang: {len(data_mentah) - len(cleaned_data)}")

# 2. Ekstraksi pelanggan unik
unique_customers = get_unique_customers(cleaned_data)
print(f"\nJumlah pelanggan unik: {len(unique_customers)}")
# Tampilkan 5 contoh ID (opsional)
print("Contoh ID pelanggan (5 pertama):", list(unique_customers)[:5])

# 3. Agregasi pendapatan per kategori
revenue_per_category = aggregate_revenue_per_category(cleaned_data)
print("\nTotal pendapatan per kategori:")
for kategori, total in revenue_per_category.items():
    print(f"  {kategori}: {total}")
```
